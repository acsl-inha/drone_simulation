```python
import numpy as np
import matplotlib.pyplot as plt

num_basic = 8
num_required = 13
num_optional = 10

basic = np.array([[1,2,3,3,3,4,4,4],[0.1,2,2.1,3.4,3.6,3.5,3.6,4]])
required = np.array([[0,0,1,1,2,3,3,3,4,4,4,5,5],[0,3,0,3,0,0.1,0,2,0,0.1,2,0,4]])
optional = np.array([[0,1,1,1,1,1,2,4,4,6],[2,0.2,1.3,0.4,4.1,4,2.3,4.1,4.2,4]])
subject_basic = np.array(["Chemistry","Physics1","Physics2","Probability and Random Variables","Calculus1","Calculus2","Differential Equation","Linear Algebra"])
subject_required = np.array(["Electromagnetics1","Basic Programming","Circuit Analysis","Object-Oriented Programming","Basic Circuit Experiments","Physical Electronics","Electromagnetics2","Signals and Systems","Logic Circuit","Electronic Circuits1","Adventure Design","Basic Circuit Analysis","Capstone Design"])
subject_optional = np.array(["Digital Signal Processing","Electronic Circuits2","Digital Communications1","Digital circuit design and language","Automatic Controls","Introduction to Machine Learning","Image Signal Processing","Robotics Engineering","Software Laboratory","Independent Advanced Studies"])

plt.figure(figsize=(32,15))
plt.scatter(basic[0,:],basic[1,:],s=100,label='Basic subject')
plt.scatter(required[0,:],required[1,:],s=100,label='Required subject')
plt.scatter(optional[0,:],optional[1,:],s=100,label='Optional subject')
for i in range(num_basic):
    plt.text(basic[0,i]+0.05,basic[1,i]-0.04,subject_basic[i],fontsize=15)
for i in range(num_required):
    plt.text(required[0,i]+0.05,required[1,i]-0.04,subject_required[i],fontsize=15)
for i in range(num_optional):
    plt.text(optional[0,i]+0.05,optional[1,i]-0.04,subject_optional[i],fontsize=15)

plt.title('Grade-Interest Graph',fontsize=30)
plt.xlabel('Grade',fontsize='xx-large')
plt.ylabel('Interest',fontsize=30)
plt.xlim(-0.5,7.2)
plt.xticks(np.arange(0,7,1),labels=['B-','B0','B+','A-','A0','A+','P'],fontsize=20)
plt.yticks(np.arange(0,5,2),labels=['low','middle','high'],fontsize=20)
plt.legend(bbox_to_anchor=(0.9, 0.55),fontsize=20)
plt.grid()
```
